#Demo Part 1
Trước khi đi vào Demo, chúng em sẽ định nghĩa lớp đối tượng Matrix và cài đặc một số thuộc tính hàm cơ bản để tiện cho việc tái sử dụng code và thiết lập các hàm theo yêu cầu của đồ án.

Lớp Matrix bao gồm:
- Tên ma trận.
- Dùng list data dùng để lưu trữ ma trận.
- Thuộc tính rows và cols dùng để lưu trữ kích thước của ma trận.
- Hàm print_matrix để hiển thị ma trận.
- Overloading toán tử để thực hiện các phép nhân, chia, cộng, trừ cho ma trận.
- Các phép biến đổi hàng và cột.
- Hàm lấy chuyển vị cho ma trận.

```python
import builtins

class Matrix:
    def __init__(self, data, name = "Unknown"): # Khởi tạo class ma trận bao gồm các thuộc tính: 1 list 2 chiều để lưu trữ ma trận, số hàng, số cột, tên
        self.data = data # list 2 chiều để lưu trữ ma trận
        self.rows = len(data) # Số hàng
        self.cols = len(data[0]) if self.rows > 0 else 0 # Số cột
        self.name = name # Tên ma trận

    def print_matrix(self): # Hàm in ma trận
        print(f"Ma trận {self.name} kích thước ({self.rows}x{self.cols}):")
        for row in self.data:
            print(row)
        print("-" * 20)

    def __repr__(self):
        return f"Matrix(name={self.name!r}, rows={self.rows}, cols={self.cols}, data={self.data!r})"

    def __add__(self, other): # Hàm cộng ma trận
        if self.rows != other.rows or self.cols != other.cols:
            raise ValueError(f"Kích thước ma trận {self.name} không khớp với {other.name} để cộng.")

        new_data = [
            [self.data[r][c] + other.data[r][c] for c in range(self.cols)]
            for r in range(self.rows)
        ]
        return Matrix(new_data, f"({self.name} + {other.name})")

    def __sub__(self, other): # Hàm trừ ma trận
        if self.rows != other.rows or self.cols != other.cols:
            raise ValueError(f"Kích thước ma trận {self.name} không khớp với {other.name} để trừ.")

        new_data = [
            [self.data[r][c] - other.data[r][c] for c in range(self.cols)]
            for r in range(self.rows)
        ]
        return Matrix(new_data, f"({self.name} - {other.name})")

    def __mul__(self, other): # Hàm nhân ma trận
        if isinstance(other, (int, float)): # Nhân ma trận cho một số thực
            new_data = [[val * other for val in row] for row in self.data]
            return Matrix(new_data, f"{self.name} * {other}")

        if isinstance(other, Matrix): # Nhân ma trận cho một ma trận khác
            if self.cols != other.rows:
                raise ValueError(f"Kích thước {self.name} không khớp với {other.name} để nhân.")

            new_data = []
            for i in range(self.rows):
                new_row = []
                for j in range(other.cols):
                    element = sum(self.data[i][k] * other.data[k][j] for k in range(self.cols))
                    new_row.append(element)
                new_data.append(new_row)
            return Matrix(new_data, f"({self.name} * {other.name})")
        raise TypeError("Chỉ hỗ trợ nhân Matrix với Matrix hoặc số thực.")

    def __rmul__(self, other):
        return self.__mul__(other)

    def copy_from(self, other): # Hàm copy ma trận
        if not self.data:
            self.data = [row[:] for row in other.data]
            self.rows = other.rows
            self.cols = other.cols
            self.name = other.name
        else:
            if self.cols != other.cols or self.rows != other.rows:
               raise ValueError("Kích thước không khớp để sao chép.")
            for i in range(other.rows):
               for j in range(other.cols):
                   self.data[i][j] = other.data[i][j]

    def copy(self, name=None):
        return Matrix([row[:] for row in self.data], self.name if name is None else name)

    def swap_rows(self, i, j): # Hàm hoán đổi hàng
        if 0 <= i < self.rows and 0 <= j < self.rows:
            self.data[i], self.data[j] = self.data[j], self.data[i]
        else:
            raise IndexError("Chỉ số hàng không hợp lệ")

    def swap_cols(self, a, b): # Hàm hoán đổi cột
        if 0 <= a < self.cols and 0 <= b < self.cols:
            for i in range(self.rows):
                self.data[i][a], self.data[i][b] = self.data[i][b], self.data[i][a]
        else:
            raise IndexError("Chỉ số cột không hợp lệ.")

    def multiply_row_with_real_number(self, i, k): # Nhân hàng i với số thực k
        if k == 0:
            raise ValueError("Số thực dùng để nhân với ma trận không được bằng 0.")

        if 0 <= i < self.rows:
            for j in range(self.cols):
                self.data[i][j] *= k
        else:
            raise IndexError("Chỉ số hàng không hợp lệ.")

    def add_multiple_of_row(self, i, j, k): # Lấy hàng thứ i + k*hàng thứ j
        if not (0 <= i < self.rows and 0 <= j < self.rows):
            raise IndexError(f"Chỉ số hàng không hợp lệ.")
        for c in range(self.cols):
            self.data[i][c] += k * self.data[j][c]

    def get_Tran(self): # Hàm lấy ma trận chuyển vị
        new_data = [[self.data[j][i] for j in range(self.rows)] for i in range(self.cols)]
        return Matrix(new_data, name=f"{self.name}^T")

builtins.Matrix = Matrix
print("Đã định nghĩa class Matrix toàn cục.")
```

#Phương pháp demo
Chúng em sẽ sử dụng các thư viện như Numpy và Sympy để viết hàm kiểm thử verify_solution(A, x, b) để chạy kiểm thử và so sánh kết quả với những hàm do chúng em tự viết. Sau đó, chúng em sẽ viết chương trình cho phép nhập vào ma trận và gọi verify_solution để in ra kết quả

```python
import numpy as np
import sympy as sp


def is_linearly_independent(vectors): # Hàm kiểm tra độc lập tuyến tính

    # Đây là hàm con để dùng trong hàm verify_solution
    if not vectors: return True
    return sp.Matrix(vectors).rank() == len(vectors)

def is_basis_simplified(base_matrix_sp, vectors):
    # Hàm kiểm tra 1 tập vector lấy từ không gian dòng hoặc cột ma trận có phải là cơ sở dòng, cột hay không

    # Đây là hàm con để dùng trong hàm verify_solution
    dim_space = base_matrix_sp.rank()

    if len(vectors) != dim_space:
        return False

    return is_linearly_independent(vectors)

def verify_solution(A, x_sol, b): # Hàm dùng thư viện Numpy và Sympy để kiểm chứng hàm của chúng em
    print(f"\n{'='*70}")
    print(f" KIỂM TRA HỆ THỐNG CHI TIẾT: MA TRẬN {A.name} ")
    print(f"{'='*70}")

    # Chuyển đổi dữ liệu
    sp_A = sp.Matrix(A.data)
    sp_b = sp.Matrix(b.data)
    np_A = np.array(A.data, dtype=float)
    np_b = np.array(b.data, dtype=float).flatten()

    # --- PHẦN 1: MA TRẬN MỞ RỘNG [A|b] SAU RREF & NGHIỆM ---
    print("\n### 1. SO SÁNH MA TRẬN MỞ RỘNG [A|b] VÀ NGHIỆM ###")

    # Tính toán RREF
    A_copy_rref = A.copy()
    b_copy_rref = b.copy()
    gaussian_eliminate(A_copy_rref, b_copy_rref)

    my_aug_rref = [A_copy_rref.data[i] + b_copy_rref.data[i] for i in range(A.rows)]
    sp_augmented = sp_A.row_join(sp_b)
    sp_aug_rref, sp_pivots = sp_augmented.rref()
    np_sp_aug_rref = np.array(sp_aug_rref.tolist(), dtype=float)
    sp_A = sp.Matrix(A.data)
    sp_RREF_A, sp_pivots = sp_A.rref()
    np_sp_A_rref = np.array(sp_RREF_A.tolist(), dtype=float)

    # 1c. So sánh ma trận A
    print(">> Ma trận A sau khi thực hiện RREF (Theo hàm do chúng em cài đặt):")
    for row in A_copy_rref.data: print(f"   {[round(v, 4) for v in row]}")

    # Sử dụng np.allclose để so sánh vì số thực có thể có sai số rất nhỏ
    a_rref_correct = np.allclose(np.array(A_copy_rref.data), np_sp_A_rref, atol=1e-5)
    print(f"--> Trạng thái RREF của A: {'ĐÚNG' if a_rref_correct else 'SAI'}")
    # Kiểm tra nghiệm Ax = b
    print(f"\n>> Kết quả nghiệm của {A.name}.x = {b.name} theo các hàm chúng em cài đặt: {x_sol}")
    # Kiểm tra tính vô nghiệm bằng SymPy

    # Hệ vô nghiệm khi hàng cuối của RREF có dạng [0 0 ... 0 | 1]
    is_sp_consistent = True
    for i in range(sp_aug_rref.rows):
        # Nếu tất cả các hệ số bên trái (A) bằng 0 nhưng cột cuối (b) khác 0
        if all(abs(sp_aug_rref[i, j]) < 1e-10 for j in range(A.cols)) and abs(sp_aug_rref[i, A.cols]) > 1e-10:
            is_sp_consistent = False
            break

    # SO SÁNH KẾT LUẬN NGHIỆM
    if not is_sp_consistent:
        # Trường hợp VÔ NGHIỆM
        my_is_no_sol = (x_sol == []) # Hàm của bạn trả về [] khi vô nghiệm
        print(f"--> Hệ phương trình VÔ NGHIỆM.")
        print(f"--> Trạng thái kết luận: {'ĐÚNG' if my_is_no_sol else 'SAI (Hàm do chúng em cài đặt vẫn cho ra nghiệm)'}")

    else:
        # Trường hợp CÓ NGHIỆM (Duy nhất hoặc Vô số)
        if len(sp_pivots) < A.cols:
            print(f"--> Hệ phương trình VÔ SỐ NGHIỆM.")
        else:
            print(f"--> Hệ phương trình CÓ NGHIỆM DUY NHẤT.")

        # Kiểm tra giá trị nghiệm bằng cách thế số
        try:
            # Tạo các giá trị thử nghiệm cho ẩn tự do t1, t2...
            symbols = sp.symbols('t1:11')
            test_cases = [{s: 0 for s in symbols}, {s: 1 for s in symbols}] # Thử t=0 và t=1

            all_tests_passed = True
            for subs_dict in test_cases:
                numeric_x = []
                for val in x_sol:
                    if isinstance(val, str):
                        # Giải biểu thức chuỗi '2.0 + 1.0*t1' thành số thực
                        numeric_x.append(float(sp.sympify(val).subs(subs_dict)))
                    else:
                        numeric_x.append(float(val))

                # Kiểm tra Ax - b = 0
                residual = np.linalg.norm(np_A @ np.array(numeric_x) - np_b)
                if residual > 1e-3:
                    all_tests_passed = False
                    break

            print(f"--> Trạng thái giá trị nghiệm: {'ĐÚNG' if all_tests_passed else 'SAI (Nghiệm không thỏa mãn Ax=b)'}")
        except Exception as e:
            print(f"--> Lỗi khi kiểm tra giá trị nghiệm: {e}")

    # --- PHẦN 2: ĐỊNH THỨC ---
    if A.rows == A.cols:
        print("\n### 2. KIỂM TRA ĐỊNH THỨC ###")
        my_det = determinant(A) #
        np_det = np.linalg.det(np_A)
        is_det_correct = abs(my_det - np_det) < 1e-4
        print(f">> Kết quả: {my_det} (theo hàm của chúng em cài đặt) | {round(np_det, 4)} (NumPy)")
        if is_det_correct:
            print(f"--> Kết quả giống nhau, hàm của chúng em cài đặt đã cho ra kết quả CHÍNH XÁC! ")
        else:
            print(f"--> Kết quả sai khác nhau quá lớn, hàm của chúng em cài đặt đã cho ra kết quả SAI!")

    # --- PHẦN 3: NGHỊCH ĐẢO ---
    print("\n### 3. KIỂM TRA NGHỊCH ĐẢO ###")
    print(">> Ma trận gốc A:")
    A.print_matrix()

    if A.rows == A.cols:
        A_inv = inverse(A) #
        if A_inv:
            print(">> Ma trận nghịch đảo S (theo hàm của chúng em cài đặt):")
            A_inv.print_matrix()

            # Kiểm tra A * A^-1 = I
            product = np_A @ np.array(A_inv.data, dtype=float)
            is_inv_correct = np.allclose(product, np.eye(A.rows), atol=1e-5)
            if is_inv_correct:
                print(f"--> Vậy A * S = I nên S là ma trận nghịch đảo của A, hàm của chúng em cài đặt đã cho ra kết quả CHÍNH XÁC!")
            else:
                print(f"--> A * S sai khác quá lớn so với I, S không phải ma trận nghịch đảo của A, hàm của chúng em cài đặt đã cho ra kết quả SAI!")
        else:
            print("-> Ma trận không có nghịch đảo.")

    # --- PHẦN 4: CƠ SỞ (BASIS) ---
    print("\n### 4. KIỂM TRA RANK & CƠ SỞ ###")
    res_basis = rank_and_basic(A) #
    sp_rank = sp_A.rank()

    rank_is_correct = (res_basis['rank'] == sp_rank)
    print(f"-> Rank: {res_basis['rank']} (Bạn) | {sp_rank} (SymPy)")
    if rank_is_correct:
         print(f"--> Kết quả giống nhau, hàm của chúng em cài đặt đã cho ra kết quả CHÍNH XÁC!")
    else:
         print(f"--> Kết quả sai khác nhau quá lớn, hàm của chúng em cài đặt đã cho ra kết quả SAI!")

    # Kiểm tra tính đúng đắn của các cơ sở bằng cách kiểm tra tính độc lập và thuộc không gian

    # Nếu các bước trước đúng, rõ ràng các danh sách các vector lấy từ hàm rank_and_basic
    # sẽ lấy từ không gian vector dòng và cột của ma trận

    # Vì thế, ta nếu các cơ sở độc lập tuyến tính và có số lượng bằng hạng ma trận
    # ta có thể kết luận hàm rank_and_basic cho ra kết quả chính xác

    print("\n[Cơ sở không gian Dòng]")
    print(f"  >> Bằng cách dùng hàm của chúng em: {res_basis['row_basis']}")
    is_row_space_ok = is_basis_simplified(sp_A, res_basis['row_basis'])
    if is_row_space_ok:
         print("Các vector thuộc không gian dòng, độc lập tuyến tính và có số lượng đúng bằng hạng ma trận!")
         print("-> Các vector trên là cơ sở dòng của ma trận")
         print("--> Vậy hàm do chúng em cài đặt đã cho ra kết quả CHÍNH XÁC!")
    else:
         print("Số lượng vector không khớp với hạng của ma trận hoặc không độc lập tuyến tính!")
         print("-> Các vector trên không là cơ sở dòng của ma trận")
         print("--> Vậy hàm do chúng em cài đặt đã cho ra kết quả SAI!")

    print("\n[Cơ sở không gian Cột]")
    print(f"  >> Bằng cách dùng hàm của chúng em: {res_basis['col_basis']}")
    col_vectors = [sp.Matrix(v).T for v in res_basis['col_basis']]
    is_col_space_ok = is_basis_simplified(sp_A, res_basis['col_basis'])
    if is_col_space_ok:
         print("Các vector thuộc không gian dòng, độc lập tuyến tính và có số lượng đúng bằng hạng ma trận!")
         print("-> Các vector trên là cơ sở cột của ma trận")
         print("--> Vậy hàm do chúng em cài đặt đã cho ra kết quả CHÍNH XÁC!")
    else:
         print("Số lượng vector không khớp với hạng của ma trận hoặc không độc lập tuyến tính!")
         print("-> Các vector trên không là là cơ sở cột của ma trận")
         print("--> Vậy hàm do chúng em cài đặt đã cho ra kết quả SAI!")


    # 4d. Kiểm tra Cơ sở không gian Nghiệm (Null Space)
    print("\n[Cơ sở không gian Nghiệm]")
    print(f"  >> Bằng cách dùng hàm của chúng em: {res_basis['null_basis']}")
    print(f"  >> SymPy:   {[list(v) for v in sp_A.nullspace()]}")
    null_correct = True
    if res_basis['rank'] < A.cols:
        if not res_basis['null_basis']:
            null_correct = False
        else:
            for v in res_basis['null_basis']:
                # Kiểm tra A * v = 0
                if np.linalg.norm(np_A @ np.array(v)) > 1e-7:
                    null_correct = False
                    break
    if null_correct:
        print(f"--> Kết quả giống nhau, hàm của chúng em cài đặt đã cho ra kết quả CHÍNH XÁC!")
    else:
        print(f"--> Kết quả sai khác nhau quá lớn, hàm của chúng em cài đặt đã cho ra kết quả SAI!")
```

```python
# --- DEMO TRƯỚC VỚI 1 MA TRẬN CỤ THỂ ---
A_test = Matrix([[10.0, 1.0, 3.0], [2.0, 3.0, 4.0],], "A_demo")
b_test = Matrix([[3.0], [7.0]], "b_demo")

# Bước 1 trong yêu cầu: Chạy thuật toán của bạn
A_copy = A_test.copy()
b_copy = b_test.copy()
gaussian_eliminate(A_copy, b_copy) # Đưa về RREF
x_sol = back_substitution(A_copy, b_copy)

# Chạy hàm Verify
verify_solution(A_test, x_sol, b_test)
```
